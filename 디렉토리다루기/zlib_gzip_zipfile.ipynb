```python
'''
데이터 전송시 대용량 데이터는 전송속도가 느리며 전송 문제가 발생할 가능성이 높다
데이터 압축의 종류: 
    무손실 압축 : 데이터 손실이 전혀 없는 압추
    손실 압축 : 사람이 눈치채지 못할 수준의 정보만 버리고 압축하는 방법
압축률: 원시 자료량(원래 데이터크기)/압축된 자료량(압축된 데이터 크기)
다양한 압축 알고리즘에 따라 압축 성능 및 시간이 좌우됨
압축: 인코딩(encoding)
압축 해제: decoding
'''
import zlib
```

```python
data = 'Life is too short. You need python.'*10000
```

```python
data
```

```python
print(len(data))
```

```python
compless_data = zlib.compress(data.encode(encoding='utf-8')) #압축할 때 인코딩을 해준 다음 해야한다
print(len(compless_data))
```

```python
compless_data
```

```python
print(f'zlib : {round(len(data)/len(compless_data), 2)}')
```

```python
org_data = zlib.decompress(compless_data).decode('utf-8')
print(len(org_data))
```

```python
org_data
```

```python
#gzip : 파일을 압축하거나 해제할 때 사용, 내부적으로 zlib 알고리즘을 사용
import gzip
```

```python
with open('org_data.txt', 'w') as f:
    f.write(data)
```

```python
with gzip.open('compressed.txt.gz', 'wb') as f:
    f.write(data.encode('utf-8'))
```

```python
with gzip.open('compressed.txt.gz', 'rb') as f:  #파일을 오픈하면서 압축도 해제한다
    org_data = f.read().decode('utf-8')
```

```python
#zipfile : 여러 개 파일을 zip 확장자로 합쳐서 압축할 때 사용하는 모듈
import zipfile
```

```python
with zipfile.ZipFile('./sample/새파일.zip', 'w') as myzip: #압축하기
    myzip.write('./sample/새파일.txt')
    myzip.write('./sample/새파일2.txt')
```

```python
with zipfile.ZipFile('./sample/새파일.zip') as myzip: #해제하기
    myzip.extractall()
```

```python
#tarfile : 여러개 파일을 tar 확장자로 합쳐서 압축할 때 사용하는 모듈
import tarfile
```

```python
with tarfile.open('./sample/새파일.tar', 'w') as mytar:
    mytar.add('./sample/새파일.txt')
    mytar.add('./sample/새파일2.txt')
```

```python
with tarfile.open('./sample/새파일.tar') as mytar:
    mytar.extractall()
```
