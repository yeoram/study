```python
import base64
```

```python
string = 'Life is too short. We need Python'
```

```python
encoded = base64.b64encode(string) #문자열은 바이트가 아니라서 오류 발생. 문자열을 아스키로 인코딩 해준 후에 사용가능
```

```python
bstring = string.encode('ascii')
```

```python
print(bstring)
```

```python
encoded = base64.b64encode(bstring)
```

```python
print(encoded)
```

```python
decoded = base64.decodebytes(encoded)
```

```python
print(decoded)
```

```python
decoded_str = decoded.decode('ascii')
print(decoded_str)
```
